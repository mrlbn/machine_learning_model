```python
def show_train(x,y,w):
    y_m = np.dot(x,w.T)
    x=x.dot(np.ones([3,1]))
    #plt.xlim(0,30)
    #plt.ylim(0,100)
    plt.scatter(x,y_m)
    plt.plot(x,y)
    plt.show()
```

```python
def gradient_descent(x,y):
    m = len(x)
    w = np.array([0,0,0])
    #终止条件
    threshold = 0.000001
    loop_max = 500000
    
    a = 0.00005
    error = np.zeros(2)
    cnt=0
    while cnt < loop_max:
        sum_m = np.array([0,0,0],dtype='float')
        for i in range(m):
            diff = (np.dot(x[i],w.T) - y[i])* x[i]
            sum_m += diff
            #print(diff)
        w = w - a/m*diff
        print(np.dot(x,w.T))
        for i in range(m):
            error[1]+=(np.dot(w.T,x[i]) - y[i])**2
        error[1]/=m
        if np.linalg.norm(error[1]-error[0])<threshold:
            break
        else:
            error[0] = error[1]
        cnt+=1
    pass
    return w

w = gradient_descent(x_train,y_train)
w
```
