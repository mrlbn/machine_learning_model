```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
x_train = np.random.randint(1,20,[5,2])
x_train
#np.dot(x_train,np.ones([2,1]))
y_train = (x_train * np.abs(np.sin(x_train))).dot(np.ones([2,1])) * 4 + np.random.randint(0,10)

x_train = np.column_stack((x_train,np.ones([5,1])))
y_train
x_train
```

```python
def show_train(x,y):
    x=x.dot(np.ones([3,1]))
    #plt.xlim(0,30)
    #plt.ylim(0,100)
    plt.scatter(x,y)
    plt.show()
    
show_train(x_train,y_train)
```
